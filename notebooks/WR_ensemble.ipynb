# Illustrates the use of the WR (Weather Regime) class for an ensemble of proxies

```python
%matplotlib inline
```

```python
import os 
import sys
import pathlib
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### import the development version of paleopy

```python
sys.path.insert(0, '../')
```

```python
from paleopy import proxy 
from paleopy import analogs
from paleopy import ensemble
from paleopy import WR
```

### determine the current working directory 

```python
cwd = pathlib.Path.cwd() 
```

```python
print(cwd)
```

#### defines the folder where the JSON files are (for the datasets) and where to save the proxy JSON files

```python
djsons = cwd.parent / 'jsons'
```

```python
pjsons = djsons / 'proxies'
```

### defines the name of the directory (contained in djsons / proxies) where the proxies constituting the ensemble are defined 

```python
ens_dir_name = 'LIANZSWP'
```

```python
ens_jsons = pjsons / ens_dir_name
```

```python
if not ens_jsons.exists(): 
    print(f"ERROR, the directory {ens_jsons} does not exists")
```

### first instantiates an `ensemble' class, passing the directory containing the proxy JSON files, and the season (to test for consistency)

```python
ens = ensemble(pjsons=ens_jsons, season='DJF')
```

### Now instantiates a `WR` class, passing the newly created ensemble object, and the classification of interest 

```python
w = WR(ens, classification='New Zealand')
```

#### plots the bar plot, significance level = 99%

### The horizontal line for each boxplot is the median (accross all the proxies), the black circle the average, and if the anomalies are significant at the level specified (argument `sig` to the `plot_bar` method) then the box (marking the interquartile range) is highlighted in black 

```python
f = w.plot_bar(sig=10)
```

### heat map with individual proxies on the y-axis, the frequency anomalies significant at the 95% level are marked by the squares, the anomalies significant at the 99% level are marked by the thick circles 

```python
f = w.plot_heatmap()
```

### same example as above but for the SW Pacific weather regimes  

```python
w = WR(ens, classification='SW Pacific')
```

```python
f = w.plot_bar(sig=10)
```
