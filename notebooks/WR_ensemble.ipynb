# Illustrates the use of the WR (Weather Regime) class for an ENSEMBLE of proxies

```python
%matplotlib inline
```

```python
import os 
import sys
import pathlib
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### import the development version of paleopy

```python
from paleopy import analogs
from paleopy import ensemble
from paleopy import WR
```

### determine the current working directory 

```python
cwd = pathlib.Path.cwd() 
```

```python
print(cwd)
```

#### defines the folder where the JSON files are (for the datasets) and where the proxy JSON files are located

```python
djsons = cwd.parent.joinpath('jsons')
```

```python
pjsons = djsons.joinpath('proxies')
```

### defines the name of the directory (contained in djsons / proxies) where the proxies constituting the ensemble are defined 

```python
ens_dir_name = 'LIANZSWP'
```

```python
ens_jsons = pjsons / ens_dir_name
```

```python
if not ens_jsons.exists(): 
    print(f"ERROR, the directory {ens_jsons} does not exists")
```

### first instantiates an `ensemble' class, passing the directory containing the proxy JSON files, and the season (this is needed to test for consistency between the proxies)

```python
ens = ensemble(pjsons=ens_jsons, season='DJF')
```

### Now instantiates a `WR` (Weather Regime) class, passing the newly created ensemble object, and the classification of interest 

```python
w = WR(ens, classification='New Zealand (Rampal et al. 2022)')
```

### plots the barplot of the regimes climatological frequencies, and the distribution of frequency anomalies accross all the proxies as boxplots

### The horizontal line for each boxplot is the median (accross all the proxies), the black circle the average, and if the anomalies are significant at the level specified in percentage (argument `sig` to the `plot_bar` method, e.g. 10 = 90%) then the box (marking the interquartile range) is highlighted in black 

```python
f = w.plot_bar(sig=10)
```

### heat map with individual proxies on the y-axis, the frequency anomalies significant at the 95% level are marked by the squares, the anomalies significant at the 99% level are marked by the thick circles 

```python
f = w.plot_heatmap()
```
